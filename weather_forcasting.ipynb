```python
import requests
import json
import csv
from requests.exceptions import RequestException
```

```python
API_KEY = "54c015c19a3fb919385d73ec246c2e87"
city = 'Islamabad'
raw_data = {}
extracted_data = {}
```

```python
try:
    response = requests.get(f"https://api.openweathermap.org/data/2.5/weather?q={city}&units=metric&appid={API_KEY}")
    if(response.status_code != 200):
        print("Error!")
        print(response.status_code)
    else:
        raw_data = response.json()
        print("API accept data")
except RequestException as RE:
    print('Request Error ',RE)

    
        
    
```

```python
try:
    extracted_data = {
        "City" : raw_data["name"],
        "Temperature" : raw_data["main"]["temp"],
        "Humidity" : raw_data["main"]["humidity"],
        "wind Speed" : raw_data["wind"]["speed"],
    }
except:
    print("Eror in extracting data")
```

```python
with open("data.json" , "a", newline="") as file:
    json.dump(extracted_data, file,indent = 4)
```

```python
with open("data.csv", "a",newline="") as file:
    
    writter = csv.DictWriter(file, fieldnames= extracted_data.keys())
    writter.writerow(extracted_data)
```
