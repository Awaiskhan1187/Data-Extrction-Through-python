```python
import requests
import json
import csv
import requests.exceptions
```

```python
API_KEY = "API_KEY"
country = "Iran"
raw_data = {}
extracted_data = {}
```

```python
url = "https://newsapi.org/v2/top-headlines"
try:
    response = requests.get(url + f"?country={country}&apiKey={API_KEY}")
    if(response.status_code != 200):
        print("API does not connect! ",response.status_code)
    else:
        raw_data = response.json()
except:
    print("Invalid request")
```

```python
key_heading = raw_data.keys()
key_heading
```

```python
for i in raw_data['articles']:
    print(i.keys())
    print("\n")
```

```python
try:
    extracted_data = {
        'Title': [],
        'Source': [],
        'Author': [],
        'Published Date': [],
        'URL': [],
    }

    for article in raw_data['articles']:
        extracted_data['Title'].append(article.get('title'))
        extracted_data['Source'].append(article.get('source'))
        extracted_data['Author'].append(article.get('author'))
        extracted_data['Published Date'].append(article.get('publishedAt'))
        extracted_data['URL'].append(article.get('url'))

except Exception as e:
    print("Extraction failed:", e)
```

```python
print(extracted_data)
```

```python
with open("news.json", "a") as file:
    json.dump(extracted_data,file,indent = 4)
```

```python
with open('news.csv',"a") as file:
    writer = csv.DictWriter(file, fieldnames = extracted_data.keys())
    writer.writerow(extracted_data)
    
```
